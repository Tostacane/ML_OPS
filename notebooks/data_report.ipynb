# Data Report

```python
import numpy as np
import pandas as pd
from src.data.loaders import load_dataset
```

```python
df = load_dataset("hr_attrition")

df = df.dropna()
df = df.reset_index(drop=True)
```

```python
def print_column_info(df):
    for col in df.columns:
        col_data = df[col]
        dtype = col_data.dtype

        print(f"\n=== {col} ===")

        # Numerica
        if np.issubdtype(dtype, np.number):
            print("Tipo: numerico")
            print(f"Min: {col_data.min()}")
            print(f"Max: {col_data.max()}")

        # Categorica / altro
        else:
            print("Tipo: categorico")
            unique_vals = col_data.dropna().unique()
            print(f"Valori possibili ({len(unique_vals)}): {list(unique_vals)}")
```

```python
print_column_info(df)
```
